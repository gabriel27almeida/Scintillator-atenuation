```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read simulation data

mu = 0.5 # cm^-1
r = 0.8
file_path = 'data/run5.txt'
skip_rows = 16

data = np.genfromtxt(file_path, skip_header=skip_rows, names=True)
x = data['x']
y = data['y']*400 # cm to 
intensity = data['intensity'] / data['intensity'][0]
```

```python
# Read experimental data

label = '50-50 PEN PET'
file_path = 'data/exp1.txt'

data_exp = np.genfromtxt(file_path, names=True)
y_exp = data_exp['y']- data_exp['y'][0]
intensity_exp = data_exp['intensity'] / data_exp['intensity'][0]
```

```python
# Plotting

plt.scatter(y, intensity, marker='o', color='b', label=f'mu={mu}, r={r}', alpha=0.6)
plt.scatter(y_exp, intensity_exp, marker='o', color='r', label=label, alpha=0.6)
plt.title('Intensity profile')
plt.xlabel('y')
plt.ylabel('Intensity')
plt.legend()
plt.show()
```
