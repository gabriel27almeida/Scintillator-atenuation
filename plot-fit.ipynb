```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read fitting data

file_path = 'OUTPUT-FIT'
skip_rows = 12

data = np.genfromtxt(file_path, skip_header=skip_rows, names=True)
mu = data['mu']
r = data['r']
chi2 = data['chi2']
```

```python
# Create a 2D colormap

# Reshape z to match the shape of x and y
num_x_values = len(np.unique(mu))
num_y_values = len(np.unique(r))
chi2_reshaped = np.array(chi2).reshape(num_x_values, num_y_values)

# Create a 2D colormap
plt.figure()
plt.contourf(np.unique(r), np.unique(mu), chi2_reshaped, cmap='viridis')
plt.colorbar(label=r'$\chi^2$')
plt.xlabel(r'$r$')
plt.ylabel(r'$\mu$ (cm$^{-1}$)')
plt.title('Fitting model')

plt.show()
```

```python
r_indices = [2,4,6]
for r_indice in r_indices:
    plt.plot(np.unique(mu), chi2_reshaped[:,r_indice], label = f'r = {np.unique(r)[r_indice]}')
plt.xlabel(r'$\mu$ (cm$^{-1}$)')
plt.ylabel(r'$\chi^2$')
plt.legend()
```

```python
mu_indices = [10,20,30]
for mu_indice in mu_indices:
    plt.plot(np.unique(r), chi2_reshaped[mu_indice], label = f'mu = {np.unique(mu)[mu_indice]}')
plt.xlabel(r'$r$')
plt.ylabel(r'$\chi^2$')
plt.legend()
```
