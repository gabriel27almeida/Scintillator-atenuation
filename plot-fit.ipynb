```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read fitting data

file_path = 'OUTPUT-FIT'
skip_rows = 12

data = np.genfromtxt(file_path, skip_header=skip_rows, names=True)
mu = data['mu']
r = data['r']
chi2 = data['chi2']
```

```python
# Create a 2D colormap

# Reshape z to match the shape of x and y
num_x_values = len(np.unique(mu))
num_y_values = len(np.unique(r))
chi2_reshaped = np.array(chi2).reshape(num_y_values, num_x_values)

# Create a 2D colormap
plt.figure()
plt.contourf(np.unique(mu), np.unique(r), chi2_reshaped, cmap='viridis')
plt.colorbar(label=r'$\chi^2$')
plt.xlabel(r'$\mu$ (cm$^{-1}$)')
plt.ylabel(r'$r$')
plt.title('Fitting model')

plt.show()
```
